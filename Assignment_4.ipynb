```python
sales = [1200,450,980,1500,3000]

#comma-separated format
with open ("sales_data.txt","w") as f:
    #for s in sales:
    #    f.write(",".join(str(s)))  #Here str(s) is just one string & join() treats it as an iterable of characters
    f.write(",".join(str(s) for s in sales))  
with open ("sales_data.txt","r") as f:
    content = f.read()
    print(content)

print("\n") 

# write each sale on a new line
with open ("sales_data.txt","w") as f:
    for s in sales:
        f.write(str(s) + "\n")
# reopen and print contents
with open ("sales_data.txt","r") as f:
    content = f.read()
    print(content)
      

```

```python
new_sales = [5000, 2500,1700]

# append new sales to existing file
with open ("sales_data.txt","a") as f:
    for s in new_sales:
        f.write(str(s) + "\n")

# reopen and print entire updated file
with open ("sales_data.txt","r") as f:
    content = f.read()
    print(content)

#print total number of lines
with open ("sales_data.txt","r") as f:
    tot_lines = len(f.readlines())
    print(tot_lines)
```

task 4

```python
with open ("sales_data.txt","r") as f:
    content = f.readlines() #read all lines into a list
    sales = [int(c.strip()) for c in content]  #convert each line to int after stripping whitespace

tot_sales = sum(sales)
print(tot_sales)

high_sale = max(sales)
print(high_sale)

low_sale = min(sales)
print(low_sale)

avg_sale = tot_sales / len(sales)
print(avg_sale)
```

```python
disc_price_list = []
prices = {
        "Mouse": 500,
        "Keyboard":800,
        "Monitor": 7000,
        "Pendrive": 400,
        "Camera": 5000          
}  

disc_per = int(input("Enter discount %: "))

for price in prices: 
   
    with open ("discount_report.txt","w") as f:
        for prdt, price in prices.items():
            disc = price * (disc_per/100)
            disc_price = price - disc
            disc_price_list.append(disc_price)
            f.write(f"{prdt} | {price} | {disc_price:.2f}\n")

with open ("discount_report.txt","r") as f:
    content = f.read()
    print(content)

with open ("discount_report.txt","r") as f: 
    lines = f.readlines()
    tot_items = len(lines)
    print("Total items:",tot_items)
    
    Avg_disc_price = sum(disc_price_list)/ len(disc_price_list)
    print("Avg discounted price:",Avg_disc_price)
```
