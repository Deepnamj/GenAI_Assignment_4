```python
# ⚠️  ORDER DEPENDENCY: Tasks 1 → 2 → 3 → 4 must be run in sequence.
# Task 1 creates sales_data.txt; Tasks 2, 3, and 4 depend on it existing.
sales = [1200,450,980,1500,3000]

#comma-separated format
with open ("sales_data.txt","w") as f:
    #for s in sales:
    #    f.write(",".join(str(s)))  #Here str(s) is just one string & join() treats it as an iterable of characters
    f.write(",".join(str(s) for s in sales))  
with open ("sales_data.txt","r") as f:
    content = f.read()
    print(content)

print("\n") 

# write each sale on a new line
with open ("sales_data.txt","w") as f:
    for s in sales:
        f.write(str(s) + "\n")
# reopen and print contents
with open ("sales_data.txt","r") as f:
    content = f.read()
    print(content)
      

```

task 3

```python
# ⚠️  Run Tasks 1 and 2 first — this cell appends to sales_data.txt.
new_sales = [5000, 2500,1700]

# append new sales to existing file
with open ("sales_data.txt","a") as f:
    for s in new_sales:
        f.write(str(s) + "\n")

# reopen and print entire updated file
with open ("sales_data.txt","r") as f:
    content = f.read()
    print(content)

#print total number of lines
with open ("sales_data.txt","r") as f:
    tot_lines = len(f.readlines())
    print(tot_lines)
```

task 4

```python
# ⚠️  Run Tasks 1, 2, and 3 first — this cell reads the fully-populated sales_data.txt.
with open ("sales_data.txt","r") as f:
    content = f.readlines() #read all lines into a list
    sales = [int(c.strip()) for c in content]  #convert each line to int after stripping whitespace

tot_sales = sum(sales)
print(tot_sales)

high_sale = max(sales)
print(high_sale)

low_sale = min(sales)
print(low_sale)

avg_sale = tot_sales / len(sales)
print(avg_sale)
```
